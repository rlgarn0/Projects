```python
#code from Josh
# function to create column with above average saleprice neighborhoods as 1
def richhouse(x):
    rich = ['SawyerW', 'Greens','Gilbert', 'NWAmes', 'Blmngtn',
        'CollgCr', 'Crawfor', 'ClearCr', 'Somerst', 'Timber', 'Veenker',
        'GrnHill', 'NoRidge', 'NridgHt', 'StoneBr']
    if x in rich:
        return 1
    else:
        return 0
```

```python
iowa['Abv_Avg_Neigh'] = iowa['Neighborhood']
iowa['Abv_Avg_Neigh'] = iowa['Abv_Avg_Neigh'].map(lambda x: richhouse(x))
```

```python
#set corr to a data frame
corr =iowa.corr()[['SalePrice']].sort_values('SalePrice')
```

```python
#filter corr data frame
corr_features = corr[(corr.SalePrice > .15) & (corr.SalePrice < 1) | (corr.SalePrice < -0.15) ]
```

```python
#set X as a list of rows from the corr_features data frame and y as target variable Saleprice
X_feature= [corr_features.index.values]
y= iowa['SalePrice'].map(np.log)
```

```python
#check X list of features
X_feature
```

```python
#copied X feature into X remove Neighborhoods, PID, Mas_Vnr_Area
X= iowa[['Kitchen_Qual_TA', 'Foundation_CBlock',
         'Kitchen_Qual_Fa', 'Lot_Frontage', 'Bsmt_Unf_SF', 'Garage_Yr_Blt', 'Central_Air_Y', 
         'Half_Bath',
        'Bsmt_Full_Bath', 'Lot_Area', 'Kitchen_Qual_Gd', 'Wood_Deck_SF',
        'Open_Porch_SF', 'BsmtFin_SF_1',
        'Fireplaces', 'TotRms_AbvGrd', 'Foundation_PConc',
        'Full_Bath', 'Year_Remod/Add', 'Year_Built', 'Abv_Avg_Neigh',
        'Garage_Cars', 'Garage_Area','Total_Bsmt_SF',
        'Gr_Liv_Area', 'Overall_Qual']]
```

```python
#train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
#check shape of training data
X_train.shape
```

```python
#check shape of test data
X_test.shape
```

```python
#scale data
ss = StandardScaler()
ss.fit(X_train)
X_train_sc = ss.transform(X_train)
X_test_sc = ss.transform(X_test)
```

```python
#models cv 5 to get rid of warning
lr = LinearRegression()
la = LassoCV(cv = 5)
rd = RidgeCV()
```

```python
cross_val_score(lr, X_train_sc, y_train, cv = 5).mean()
```

```python
cross_val_score(la, X_train_sc, y_train, cv = 5).mean()
```

```python
cross_val_score(rd, X_train_sc, y_train, cv = 5).mean()
```

```python
la.fit(X_train_sc, y_train)
```

```python
la.score(X_train_sc, y_train)
```

```python
#RMSE for LASSO train
la_pred_train=np.exp(la.predict(X_train_sc))
exp_y_train=np.exp(y_train)
(metrics.mean_squared_error(exp_y_train, la_pred_train))**(.5)
```

```python
#R2 LASSO Test
la.score(X_test_sc, y_test)
```

```python
#RMSE for LASSO test
la_pred_test=np.exp(la.predict(X_test_sc))
exp_y_test=np.exp(y_test)
(metrics.mean_squared_error(exp_y_test, la_pred_test))**(.5)
```

```python
#from Temple
# RMSE for the avg sale price. Null model score.
from sklearn.dummy import DummyRegressor
dummy_mean = DummyRegressor(strategy='mean')
dummy_mean.fit(X_train_sc, y_train)
dum_pred_train=np.exp(dummy_mean.predict(X_train_sc))
exp_y_train=np.exp(y_train)
(metrics.mean_squared_error(exp_y_train, dum_pred_train))**(.5)
```
