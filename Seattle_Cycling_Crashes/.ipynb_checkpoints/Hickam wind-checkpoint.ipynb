```python
import pandas as pd
import numpy as np
from datetime import datetime
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from keras.models import Sequential
from keras.layers import Dense, Dropout, GRU
from keras.optimizers import Adam
from keras.preprocessing.sequence import TimeseriesGenerator
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
%matplotlib inline
phnl_10 = pd.read_csv('PHIK_2010.csv')
phnl_10.head()
```

```python
phnl_10_c = phnl_10.drop(['relh','skyl1', 'skyl2', 'skyl3','ice_accretion_1hr', 'ice_accretion_3hr', 'ice_accretion_6hr', 'feel', 'metar'], axis =1)
```

```python
#from https://stackoverflow.com/questions/19632075/how-to-read-file-with-space-separated-values-in-pandas
nor_buoy = pd.read_csv("north_2010.csv", delimiter=r"\s+")
nor_buoy = nor_buoy.drop([0], axis = 0)
nor_buoy.rename(columns= {'#YY':'YY'}, inplace = True)
nor_buoy['date'] = pd.to_datetime(dict(year=nor_buoy.YY,
                                      month=nor_buoy.MM, 
                                       day=nor_buoy.DD, 
                                       hour=nor_buoy.hh, 
                                      minute=nor_buoy.mm))
                                 
nor_buoy.drop(['YY','MM','DD', 'hh', 'mm', 'WVHT','DPD', 'APD', 'MWD', 'WTMP', 'VIS','TIDE', 'DEWP'], axis =1, inplace = True)
#from https://stackoverflow.com/questions/15891038/change-data-type-of-columns-in-pandas               
nor_buoy[['WDIR','WSPD','GST','PRES','ATMP']] = nor_buoy[['WDIR','WSPD','GST','PRES','ATMP']].apply(pd.to_numeric)
nor_buoy.rename(str.lower, axis='columns', inplace = True)
nor_buoy['nor_wdir']=nor_buoy['wdir']
nor_buoy['nor_wspd_kts'] = nor_buoy['wspd'] * 1.944
#nor_buoy['nor_gust_kts'] = nor_buoy['gst'] * 1.944
nor_buoy['nor_pres_inhg'] = np.round((nor_buoy['pres'] * 0.029529983071445),2)
nor_buoy.drop(['wdir','wspd','gst', 'pres','atmp'], axis =1, inplace = True)
```

```python
sou_buoy = pd.read_csv("south_2010.csv", delimiter=r"\s+")
sou_buoy = sou_buoy.drop([0], axis = 0)
sou_buoy.rename(columns= {'#YY':'YY'}, inplace = True)
sou_buoy['date'] = pd.to_datetime(dict(year=sou_buoy.YY,
                                      month=sou_buoy.MM, 
                                       day=sou_buoy.DD, 
                                       hour=sou_buoy.hh, 
                                      minute=sou_buoy.mm))
                                 
sou_buoy.drop(['YY','MM','DD', 'hh', 'mm', 'WVHT','DPD', 'APD', 'MWD', 'WTMP', 'VIS','TIDE', 'DEWP'], axis =1, inplace = True)
#from https://stackoverflow.com/questions/15891038/change-data-type-of-columns-in-pandas               
sou_buoy[['WDIR','WSPD','GST','PRES','ATMP']] = sou_buoy[['WDIR','WSPD','GST','PRES','ATMP']].apply(pd.to_numeric)
sou_buoy.rename(str.lower, axis='columns', inplace = True)
sou_buoy['sou_wdir']=sou_buoy['wdir']
sou_buoy['sou_wspd_kts'] = sou_buoy['wspd'] * 1.944
#sou_buoy['sou_gust_kts'] = sou_buoy['gst'] * 1.944
sou_buoy['sou_pres_inhg'] = np.round((sou_buoy['pres'] * 0.029529983071445),2)
sou_buoy.drop(['wdir','wspd','gst', 'pres','atmp'], axis =1, inplace = True)
```

```python
phnl_10_test = phnl_10_c[['valid', 'tmpf', 'drct', 'sknt', 'alti', 'vsby', 'gust']]
```

```python

phnl_10_test.replace({'M': 0}, inplace =True)
```

```python
var =[]
for item in phnl_10_test['gust'].items():
    if item[1] != 0.0:
         var.append(item[1])
    else:
        var.append(phnl_10_test['sknt'].iloc[item[0]])
#add to a dataframe
phnl_10_test['gust']= pd.DataFrame({'gust': var}, columns =['gust'])
```

```python
phnl_10_test.head()
```

```python
#from https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.to_datetime.html
phnl_10_test['date'] = pd.to_datetime(phnl_10_test.valid ,infer_datetime_format=True)
```

```python
phnl_10_test[['tmpf','drct','sknt','alti','vsby','gust']] = phnl_10_test[['tmpf','drct','sknt','alti','vsby','gust']].apply(pd.to_numeric)
```

```python
# converting this to the index so we can preserve the date_start_time columns so you can validate the merging logic
phnl_10_test.index = phnl_10_test['date']
nor_buoy.index = nor_buoy['date']
sou_buoy.index = sou_buoy['date']
# the magic happens below, check the direction and tolerance arguments
tol = pd.Timedelta('29 minute')
comb_2010 = pd.merge_asof(left=phnl_10_test,right=nor_buoy,right_index=True,left_index=True,direction='nearest',tolerance=tol)
comb_2010 = pd.merge_asof(left=comb_2010,right=sou_buoy,right_index=True,left_index=True,direction='nearest',tolerance=tol)

```

```python
comb_2010.dropna(inplace = True)

comb_2010['diff_buoy'] = comb_2010['nor_pres_inhg']-comb_2010['sou_pres_inhg']

comb_2010['diff_nor'] = comb_2010['nor_pres_inhg']-comb_2010['alti']

comb_2010['diff_sou'] = comb_2010['alti']-comb_2010['sou_pres_inhg']

comb_2010.drop(['date_x', 'date_y', 'date', 'valid', 'sknt'], axis =1, inplace= True)

comb_2010= comb_2010[comb_2010['nor_pres_inhg'] <31]

comb_2010= comb_2010[comb_2010['sou_pres_inhg'] <31]

comb_2010['sq_diff_buoy'] = comb_2010['diff_buoy']**2

comb_2010['sq_diff_nor'] = comb_2010['diff_nor']**2

comb_2010['sq_diff_sou'] = comb_2010['diff_sou']**2
comb_2010.index = pd.date_range('1/1/2010', 
                               periods=len(comb_2010.index), # we want one period per entry
                               freq='H')
```

```python
comb_2010[comb_2010['gust']>24]
```

```python
# Create first_diff_ridership and second_diff_ridership
# from lesson 9.03-9.04
comb_2010['first_diff_gust'] = comb_2010['gust'].diff(1)
comb_2010['second_diff_gust'] = comb_2010['gust'].diff(1).diff(1)
```

```python
# Import Augmented Dickey-Fuller test.
#from lesson 9.03-9.04
from statsmodels.tsa.stattools import adfuller
# Code written by Joseph Nelson.

def interpret_dftest(dftest):
    dfoutput = pd.Series(dftest[0:2], index=['Test Statistic','p-value'])
    return dfoutput
```

```python
# Run ADF test on original (non-differenced!) data.
interpret_dftest(adfuller(comb_2010['gust']))
```

```python
# Run the ADF test on our once-differenced data.
interpret_dftest(adfuller(comb_2010['first_diff_gust'].dropna()))
```

```python
# Run the ADF test on our once-differenced data.
interpret_dftest(adfuller(comb_2010['second_diff_gust'].dropna()))
```

```python
# Generate plot.
plot_acf(comb_2010['gust'].dropna(), lags = 30);

```

```python
# Generate plot.
plot_pacf(comb_2010['gust'].dropna(), lags = 30);
```

```python
plot_pacf(comb_2010['first_diff_gust'].dropna(), lags = 30);
```

```python
X = comb_2010.drop(['gust', 'first_diff_gust', 'second_diff_gust'], axis =1)
y= comb_2010['gust']

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size = 0.3,
                                                    shuffle = False) # Shuffle = False is IMPORTA
X_train.head()
```

```python
# Convert training/testing sets to a DataFrame.
# ss = StandardScaler()
# X_train = ss.fit_transform(X_train)
# X_test = ss.transform(X_test)
X_train = pd.DataFrame(X_train)
X_test = pd.DataFrame(X_test)

# Add column for lagged values of precipitation.
X_train_shift = X_train.shift(1)
X_test_shift = X_test.shift(1)

# Check out top of the test dataframe.
X_test_shift.shape
```

```python
X_test_shift.index[1]
```

```python
# Import SARIMAX
from statsmodels.tsa.statespace.sarimax import SARIMAX
sarima = SARIMAX(endog = y_train[1:],
                 order = (1, 0, 1),              # (p, d, q)
                 seasonal_order = (1, 0, 0, 24), # (P, D, Q, S)
                 exog = X_train_shift.dropna()) 

# Fit SARIMA model.
model = sarima.fit(maxiter=1000)
```

```python
# Generate predictions based on test set.
preds = model.predict(start=X_test_shift.index[1],
                      end=X_test_shift.index[-2],
                      exog=X_test_shift.dropna())

# Evaluate predictions.
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test[1:len(y_test)-1], preds))

# Plot data.
plt.figure(figsize=(10,6))
plt.plot(y_train, color = 'blue')
plt.plot(y_test, color = 'orange')
plt.plot(preds, color = 'green')
plt.title(label = 'Predictions', fontsize=16)
plt.show();
```

```python
# Starting MSE and (P, D, Q, S).
best_mse = 99 * (10 ** 16)
final_P = 0
final_D = 0
final_Q = 0
final_S = 0

for P in range(3):
    for Q in range(3):
        for D in range(3):
            for S in range(0,24,6):
                try:
                    print(f'Attempting to fit SARIMAX(0,2,1)x({P},{D},{Q},{S})')
                    # Instantiate SARIMAX model.
                    sarimax = SARIMAX(endog = y_train[1:],
                                     order = (1, 0, 1),              # (p, d, q)
                                     seasonal_order = (P, D, Q, S),  # (P, D, Q, S)
                                     exog = X_train_shift.dropna()) 
 
                    # Fit SARIMAX model.
                    model = sarimax.fit(maxiter=1000)

                    # Generate predictions based on test set.
                    preds = model.predict(start=X_test_shift.index[1],
                                          end=X_test_shift.index[-2],
                                          exog=X_test_shift.dropna())
                    # Evaluate predictions.
                    print(f'The MSE for (0, 2, 1)x({P},{D},{Q},{S}) is: {mean_squared_error(y_test[1:len(y_test)-1], preds)}')

                    # Save for final report.
                    if best_mse > mean_squared_error(y_test[1:len(y_test)-1], preds):
                        best_mse = mean_squared_error(y_test[1:len(y_test)-1], preds)
                        final_P = P
                        final_D = D
                        final_Q = Q
                        final_S = S

                except:
                    pass

print(f'Our model that minimizes MSE on the testing data is the SARIMAX(0, 2, 1)x({final_P},{final_D},{final_Q},{final_S}).')
print(f'This model has an MSE of {best_mse}.')
```
